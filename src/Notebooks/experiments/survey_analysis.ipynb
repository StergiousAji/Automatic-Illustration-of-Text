```python
import pandas as pd
import numpy as np

usability_df = pd.read_csv("survey_responses.csv")
usability_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style({'axes.grid' : True})

ratings = usability_df['How intuitive and easy to use did you find the application?'].values
print(f"Mean usability rating = {ratings.mean():.2f} | std={ratings.std():.2f}")

distinct_ratings = np.unique(ratings, return_counts=True)
rating_counts = np.zeros(5)
rating_counts[distinct_ratings[0]-1] = distinct_ratings[1]

df = pd.DataFrame({"rating": range(1, 6), "counts": rating_counts})

fig = plt.figure(figsize=(6, 4), dpi=500)
ax = sns.barplot(data=df, x="rating", y="counts")
ax.set_title("USABILITY RATING DISTRIBUTION\n", fontweight="bold")
ax.set_xlabel("\nUsability Rating", fontweight="bold")
ax.set_ylabel("Number of Participants", fontweight="bold")
ax.set_yticks(range(0, 10))
ax.set_ylim(top=10)
ax.set_axisbelow(True)
# ax.bar_label(ax.containers[0])
```
